```python
import numpy as np
import matplotlib.pyplot as plt
import math

from skimage import io
from skimage.color import rgb2hsv, rgb2lab
```

Loading Image

```python
direction = '../Klebeverbindungen_Daten/ProbeE2/2D/'
files = ['ProbeE2_1_makro_cut.png', 'ProbeE2_2_makro_cut.png']

images = []

for _file in files:
    images.append(io.imread(direction + _file))

fig, (ax0, ax1) = plt.subplots(ncols= 2, figsize=(8,2))

ax0.imshow(images[0])
ax0.set_title("ProbeE2_1")
ax0.axis('off')

ax1.imshow(images[1])
ax1.set_title("ProbeE2_2")
ax1.axis('off')

fig.tight_layout()
```

```python
for image in images:
    _ = plt.hist(image.ravel(), bins = 256, color = 'orange', )
    _ = plt.hist(image[:, :, 0].ravel(), bins = 256, color = 'red', alpha = 0.5)
    _ = plt.hist(image[:, :, 1].ravel(), bins = 256, color = 'Green', alpha = 0.5)
    _ = plt.hist(image[:, :, 2].ravel(), bins = 256, color = 'Blue', alpha = 0.5)
    _ = plt.xlabel('Intensity Value')
    _ = plt.ylabel('Count')
    _ = plt.legend(['Total', 'Red_Channel', 'Green_Channel', 'Blue_Channel'])
    plt.show()
```

RGB to LAB

```python
lab_images = []
for image in images:
    lab_images.append(rgb2lab(image))
    

fig, (ax0, ax1) = plt.subplots(ncols= 2, figsize=(8,2))

ax0.imshow(lab_images[0])
ax0.set_title("ProbeE2_1_LAB")
ax0.axis('off')

ax1.imshow(lab_images[1])
ax1.set_title("ProbeE2_2_LAB")
ax1.axis('off')

fig.tight_layout()
```

```python
for image in lab_images:
    _ = plt.hist(image.ravel(), bins = 512, color = 'orange', )
    _ = plt.hist(image[:, :, 0].ravel(), bins = 512, color = 'red', alpha = 0.5)
    _ = plt.hist(image[:, :, 1].ravel(), bins = 512, color = 'Green', alpha = 0.5)
    _ = plt.hist(image[:, :, 2].ravel(), bins = 512, color = 'Blue', alpha = 0.5)
    _ = plt.xlabel('Intensity Value')
    _ = plt.ylabel('Count')
    _ = plt.legend(['Total', 'L_Chanel', 'A_Channel', 'B_Channel'])
    plt.show()
```

RGB to HSV

```python
hsv_images = []
for image in images:
    hsv_images.append(rgb2hsv(image))

fig, (ax0, ax1) = plt.subplots(ncols= 2, figsize=(8,2))

ax0.imshow(hsv_images[0])
ax0.set_title("ProbeE2_1_HSV")
ax0.axis('off')

ax1.imshow(hsv_images[1])
ax1.set_title("ProbeE2_2_HSV")
ax1.axis('off')

fig.tight_layout()
```

```python
for image in hsv_images:
    _ = plt.hist(image.ravel(), bins = 512, color = 'orange', )
    _ = plt.hist(image[:, :, 0].ravel(), bins = 512, color = 'red', alpha = 0.5)
    _ = plt.hist(image[:, :, 1].ravel(), bins = 512, color = 'Green', alpha = 0.5)
    _ = plt.hist(image[:, :, 2].ravel(), bins = 512, color = 'Blue', alpha = 0.5)
    _ = plt.xlabel('Intensity Value')
    _ = plt.ylabel('Count')
    _ = plt.legend(['Total', 'H_Channel', 'S_Channel', 'V_Channel'])
    plt.show()
```

RGB to OPP 

```python
def rgb2opp(image):
    opp_image = np.zeros(image.shape,dtype=np.float32)
    r = image[:, :,0]
    g = image[:, :,1]
    b = image[:, :,2]

    opp_image[:,:,0] = (r-g) / math.sqrt(2)
    opp_image[:,:,1] = (r+g - 2*b) / math.sqrt(6)
    opp_image[:,:,2] = (r+g+b) / math.sqrt(3)

    return opp_image
```

```python
opp_images = []
for image in images:
    opp_images.append(rgb2opp(image))

fig, (ax0, ax1) = plt.subplots(ncols= 2, figsize=(8,2))

ax0.imshow(opp_images[0])
ax0.set_title("ProbeE2_1_OPP")
ax0.axis('off')

ax1.imshow(opp_images[1])
ax1.set_title("ProbeE2_2_OPP")
ax1.axis('off')

fig.tight_layout()
```

Showing Histograms
